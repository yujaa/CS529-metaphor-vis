```python
# Load libraries
import pandas as pd
from nltk.parse import CoreNLPParser
```

```python
#POS tagger from Core NLP Parser
pos_tagger = CoreNLPParser(url='http://localhost:9000', tagtype='pos')
```

```python
data = pd.read_csv('train_visualization_data.csv')
data = data.loc[:, ~data.columns.str.contains('^Unnamed')]
data.head()
```

```python
data['sentence_txt'].head()
```

```python
data['w_pos'] = 'None'
data['m_pos'] = 'None'
data.head()
```

```python
columns = ['sentence_id','word', 'metaphor', 'Score', 'sentence_txt']
error_data = pd.DataFrame(columns=columns)
error_data = error_data.fillna(0)
error_data.head()
```

```python
for index, row in data.iterrows():
    word = data.loc[index, 'word']
    metaphor = data.loc[index, 'metaphor']
    sentence = data.loc[index, 'sentence_txt']
    try:
        sentence = sentence.replace(word, " "+word)
        sentence = sentence.replace(metaphor, " "+metaphor)
        tags = dict(pos_tagger.tag(sentence.split()))
        data.loc[index, 'w_pos'] = tags[word]
        data.loc[index, 'm_pos'] = tags[metaphor]
    except:
        print(sentence)
        error_data.loc[data.index[index]] = data.iloc[index]  
```

```python
data.head()
```

```python
error_data.head()
```

```python
#Apply POS tag names to word pos
data['w_pos'] = data['w_pos'].replace(['NN', 'NNS'], 'Noun')
data['w_pos'] = data['w_pos'].replace(['NNP', 'NNPS', 'PRP'], 'Proper Noun')
data['w_pos'] = data['w_pos'].replace(['VB','VBD','VBG','VBN','VBP','VBZ'], 'Verb')
data['w_pos'] = data['w_pos'].replace(['RB','RBR','RBS','WRB'], 'Adverb')
data['w_pos'] = data['w_pos'].replace(['JJ','JJR','JJS'], 'Adjective')
#Apply POS tag names to metaphor pos
data['m_pos'] = data['m_pos'].replace(['NN', 'NNS'], 'Noun')
data['m_pos'] = data['m_pos'].replace(['NNP', 'NNPS', 'PRP'], 'Proper Noun')
data['m_pos'] = data['m_pos'].replace(['VB','VBD','VBG','VBN','VBP','VBZ'], 'Verb')
data['m_pos'] = data['m_pos'].replace(['RB','RBR','RBS','WRB'], 'Adverb')
data['m_pos'] = data['m_pos'].replace(['JJ','JJR','JJS'], 'Adjective')
```

```python
data.head()
```

```python
#Delete other rows with error tags or other tags - word POS
data = data[data.w_pos != 'None']
data = data[(data.w_pos == 'Noun') | (data.w_pos == 'Proper Noun') | (data.w_pos == 'Verb') | (data.w_pos == 'Adverb') | (data.w_pos == 'Adjective')]

#Delete other rows with error tags or other tags - metaphor POS
data = data[data.m_pos != 'None']
data = data[(data.m_pos == 'Noun') | (data.m_pos == 'Proper Noun') | (data.m_pos == 'Verb') | (data.m_pos == 'Adverb') | (data.m_pos == 'Adjective')]
```

```python
#Rename Columns
data.rename(columns = {'sentence_id':'ID', 'word':'Word', 'metaphor':'Metaphor', 'sentence_txt':'Sentence',
                       'w_pos':'Word POS Tag', 'm_pos':'Metaphor POS Tag'}, inplace = True) 
data = data[['ID', 'Word', 'Metaphor', 'Score', 'Word POS Tag', 'Metaphor POS Tag', 'Sentence']]
data.head()
```

```python
error_data.to_csv("error_data.csv")
data.to_csv("train_data.csv")
```
